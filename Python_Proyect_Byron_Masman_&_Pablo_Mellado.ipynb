#### Part 1. let´s define an equidistant grid from t= 0 to t= 2 with 51 grid points.

```python
t_start = 0
t_end = 2
grid_points = 51

delta_time = (t_end-t_start)/grid_points
grid_list = [n * delta_time for n in range(grid_points+1)]
```

At this point, we generate different values for t (time) in order to create a grid of values ranging from 0 to 2, consisting of 51 values that will be stored in a list 

#### Part 2. Calculate $\cos(2\pi t)$ and $\sin(2\pi t)$ using the euler relations. Plot both functions on the same graph and explain that represent the standar sine and cosine graphs

```python

import matplotlib.pyplot as plt
pi = 3.1415 
e = 2.7183 #Values obtained from the task statement

cos = []
sin = []

for t in grid_list:
    euler_relation = e**(1j * 1 * pi * t)
    
    #Euler formula has a real part that is the cosine and an imaginary part that is the sine, given the nature of the relationship 
    cos.append(euler_relation.real)
    sin.append(euler_relation.imag)

plt.figure(figsize=(10, 5))
plt.plot(grid_list, cos, label='cos(2πt) (Real)', color='blue')
plt.plot(grid_list, sin, label='sin(2πt) (Imaginary)', color='red')

plt.title("sine and cosine graph ")
plt.xlabel("Time t")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

The exercise shows that the complex exponential $e^{ix}$ represents a point rotating in the complex plane, forming a unit circle. Its horizontal coordinate corresponds to $\cos(x)$, while its vertical coordinate corresponds to $\sin(x)$.

This relationship follows from Euler’s formula:

$$
e^{ix} = \cos(x) + i\sin(x)
$$

When these components are plotted, the real part (cosine) and the imaginary part (sine) trace out a circular path, clearly forming the unit circle.

### Part 3. Calculate x(t) and y(t) for the canonical triplet A=1, B=1, and C = 0. Plot against y to show that it does indeed form a spiral.

```python
# first we define the parameters
A = 1
B = 1
C = 0
# We created a list to store x and y axis values
X = []
Y = []
# Now, we use the Euler´s formula 
for t in grid_list:
    z = (A*t) * e**(1j*2*pi*B*(t+C))

    X.append(z.real)
    Y.append(z.imag)
# Now we graph the results stored in "X" and "Y".
plt.figure(figsize = (6,6))
plt.plot(X,Y,label = "espiral (A=1, B=1, C=0)", color = "blue")
plt.title("Spiral Graph")
plt.xlabel("x axis (Real)")
plt.ylabel("y axis (imaginary)")
plt.axhline(0, color ="black", linewidth = 0.5)
plt.axvline(0, color = "black", linewidth = 0.5)
plt.grid(True)
plt.legend()
plt.show()

```

The spiral happens because the exponential makes the point rotate, but the $(A \cdot t)$ factor makes the radius grow at the same time.
